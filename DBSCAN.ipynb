```python
import pandas as pd
import matplotlib.pyplot as plt

# load dataset
df = pd.read_csv("cluster_moons.csv")

# check first rows
print(df.head())

# scatter plot
plt.figure(figsize=(6,6))
plt.scatter(df['X1'], df['X2'])

plt.title("Moons Dataset (Raw Data)")
plt.xlabel("X1")
plt.ylabel("X2")

plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(df['X1'], df['X2'], s=40, edgecolor='k')

plt.title("Moons Dataset Visualization")
plt.xlabel("X1")
plt.ylabel("X2")
plt.grid(alpha=0.3)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
df = pd.read_csv("cluster_moons.csv")
X = df[['X1','X2']]
db = DBSCAN(eps=0.2, min_samples=5)
labels = db.fit_predict(X)
df['cluster'] = labels
print(df['cluster'].value_counts())
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df[df.cluster != -1]['X1'],
            df[df.cluster != -1]['X2'],
            c=df[df.cluster != -1]['cluster'],
            cmap='rainbow',
            s=40,
            label="Clusters")
plt.scatter(df[df.cluster == -1]['X1'],
            df[df.cluster == -1]['X2'],
            c='black',
            s=60,
            marker='x',
            label="Outliers")
plt.title("DBSCAN Clustering with Outliers")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.show()
```

```python
outliers = df[df.cluster == -1]
print("Number of outliers:", len(outliers))
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt
X = df[['X1','X2']].values
neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(X)
distances, indices = neighbors_fit.kneighbors(X)
distances = np.sort(distances[:,4])
plt.figure(figsize=(7,5))
plt.plot(distances)
plt.title("K-Distance Graph (Choose eps at elbow)")
plt.xlabel("Points sorted by distance")
plt.ylabel("5th Nearest Neighbor Distance")
plt.show()
```
